##Exploratory Data Analysis (EDA)

Exploratory Data Analysis refers to the critical process of performing initial investigations on data so as to discover patterns,to spot anomalies,to test hypothesis and to check assumptions with the help of summary statistics and graphical representations.

##Total sales, total quantity sold and total profit

```python
# total Sales
print("Total Sales: ", round(sum(df['Sales']), 2))

# total Quantity sold
print("Total quantity sold: ", sum(df['Quantity']))

# total Profit
print("Total Profit: ", round(sum(df['Profit']), 2))
```

## What are the top 10 cities that generate the most profit

```python
cities_profit = pd.DataFrame(df.groupby('City')['Profit'].sum())
cities_profit.reset_index(inplace=True)
```

```python
cities_profit = cities_profit.sort_values( by="Profit",ascending=False)
cities_profit.head(10)
```

```python
import plotly.express as px
fig = px.treemap(cities_profit,
                 path=['City' ,'Profit'],
                 color_continuous_scale='deep',
                 values='Profit',color='Profit')

fig.update_layout(width=1000, height=500)
fig.show()
```

## What are the top 10 states that generate the most profit

```python
states_profit = pd.DataFrame(df.groupby('State')['Profit'].sum())
states_profit.reset_index(inplace=True)
states_profit = states_profit.sort_values( by="Profit",ascending=False)
states_profit.head(10)
```

```python
import plotly.express as px
fig = px.treemap(states_profit,
                 path=['State' ,'Profit'],
                 color_continuous_scale='deep',
                 values='Profit',color='Profit')
fig.update_layout(width=1000, height=500)
fig.show()
```

The top 3 most profitable states are California, New York, and Washington, with profits of $$76,330.79, $74,015.46, and $33,368.24, respectively.

###How does Sales, Quantity and Discount affects SuperStores Profits?

```python
# compare linear relationships between attributes using correlation coefficient generated using
# correlation matrix
sns.heatmap(df.corr(), cmap = 'PuBu', annot = True)
plt.show()
```

We can clearly see that "Sales" and "Discount" have a direct impact on our "Profit," with Sales showing a positive correlation and Discount showing a negative correlation ,respectively. This is expected because as sales rise, profits rise as well, and as discounts are offered on sales, profit margins fall.

```python
fig = px.scatter(df,x="Profit",y="Sales",color="Discount",
                 size="Quantity",symbol="Segment",title="How diffrent factors affects Superstore's sales ")
fig.update_layout(height=600, width=800,
                  legend=dict(yanchor="top", y=0.99,
                              xanchor="left", x=0.01))
fig.show()
```

The fact that higher sales with lower discounts resulted in a lower loss than lower sales with larger discounts indicates that discounting plays a significant role in the Superstore's loss.

The scatter plot shows the same picture as the heat map, indicating that the majority of profitable sales had discounts of less than 40%.
